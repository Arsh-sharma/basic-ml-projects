```python
ai21_51=df["Annual Income(k$)"][(df["Annual Income(k$)"]>=21) & (df["Annual Income(k$)"]<=51)]
ai52_72=df["Annual Income(k$)"][(df["Annual Income(k$)"]>=52) & (df["Annual Income(k$)"]<=72)]
ai73_103=df["Annual Income(k$)"][(df["Annual Income(k$)"]>=73) & (df["Annual Income(k$)"]<=103)]
ai104_124=df["Annual Income(k$)"][(df["Annual Income(k$)"]>=104) & (df["Annual Income(k$)"]<=124)]
ai125_149=df["Annual Income(k$)"][(df["Annual Income(k$)"]>=125) & (df["Annual Income(k$)"]<=149)]

aix=["$(21-51)k","$(52-72)k","$(73-103)k","$(104-124)k","$(125-149)k"]
aiy=[len(ai21_51.values),len(ai52_72.values),len(ai73_103.values),len(ai104_124.values),len(ai125_149.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=aix,y=aiy,palette='Spectral')
plt.title("Annual incomes")
plt.xlabel("income")
plt.ylabel("NO. OF CUSTOMERS")
plt.show()
```

```python
X1=df.loc[:,["Age","Spending Score(1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
label=kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0],X1[:,1], c=kmeans.labels_,cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black")
plt.title("Cluster of Customers")
plt.xlabel("Age")
plt.ylabel("Spending Score(1-100)")
plt.show()
```
